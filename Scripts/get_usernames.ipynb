```python
import json
import os
```

```python
folder_path = r'/Users/ananthu/Desktop/venv/cheater accounts'
```

```python
test_path = r'/Users/ananthu/Desktop/venv/records_1_to_10.json'
```

```python
cheater_list = []
for file in os.listdir(folder_path):
    path_file = os.path.join(folder_path, file)
    with open(path_file, 'r') as file:
        data = json.load(file)
    for i in range(1,20):
        dat = data[i]
        name = dat['game_account']['display_name']
        cheater_list.append(name)
        
```

```python
# Save as a text file
with open('cheater_list.txt', 'w') as file:
    for item in cheater_list:
        file.write(f"{item}\n")
```
